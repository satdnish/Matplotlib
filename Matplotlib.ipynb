```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.arange(10,21)
x
```

```python
y=2*x
y
```

```python
y2=3*x
y2
```

```python
plt.plot(x,y,linewidth=2,linestyle='-.',color='r')
plt.grid()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line plot')
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

# subplot

```python
plt.subplot(2,1,1)# 2=row,1=column,1=index
plt.plot(x,y)

plt.subplot(2,1,2)
plt.plot(x,y2)
plt.show()
```

# Barplot

we used barplot for categorial understand

```python
student={'Bob':87,'Matt':56,'Sam':27}
```

```python
name=list(student.keys())
values=list(student.values())
```

```python
plt.bar(name,values,color='r')
plt.title('Verticle bar plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
plt.barh(name,values,color='g')
plt.title('Horizontal bar plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

# Scatter plot

```python
x=[1,3,4,2,4,6,6,7,9]
```

```python
y=[9,7,5,8,4,9,3,2,5]
```

```python
z=[5,6,4,5,6,7,3,9,0]
```

```python
plt.scatter(x,y,marker='*',color='b',s=100)# marker for style, s means size
plt.show()
```

2 scatter in 1 plot

```python
plt.scatter(x,y,marker='*',color='b',s=100)
plt.scatter(x,z,marker='*',color='r',s=100)
plt.title('2 scatter in 1 plot')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.scatter(x,y,marker='*',color='b',s=100)
plt.title('scatter subplot')

plt.subplot(2,1,2)
plt.scatter(x,z,marker='*',color='r',s=100)
plt.show()
```

# Pie-Chart 

Genrally used for disterbution show

```python
fruits=['Mango','Apple','Banana','Pineapple']
```

```python
quantities=[50,100,150,80]
```

```python
plt.pie(quantities,labels=fruits,autopct='%0.1f%%',colors=['red','yellow','green','blue'])#autopct='%0.1f%%' for percent attribute 0.1 mean how much value wanna after .1,2 or 3....n
plt.show()
```

# DoughNut-Char

```python
plt.pie(quantities,labels=fruits,autopct='%0.1f%%',radius=1) #radius attribute for inner and outer edge of circle
plt.pie([100],radius=0.5,colors='w')
plt.show()
```
